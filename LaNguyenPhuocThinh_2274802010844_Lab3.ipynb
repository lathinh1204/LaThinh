```python
import pandas as pd
import numpy as np
import unicodedata
import re
from google.colab import files

print("=" * 80)
print("UPLOAD FILE CSV VĂO GOOGLE COLAB")
print("=" * 80)

uploaded = files.upload()
csv_file = list(uploaded.keys())[0]

print(f"\n/ File ÄĂŁ upload: {csv_file}")

column_names = ["Id", "Name", "Age", "Weight", "m0006", "m0612", "m1218", "f0006", "f1218"]

df = pd.read_csv(
    csv_file,
    names=column_names,
    engine='python',
    on_bad_lines='skip',
    encoding='utf-8-sig',
    index_col=False  # Add this to prevent the first column from being used as index
)

print("\n" + "=" * 80)
print("PROBLEM 1: Load dáťŻ liáťu táťŤ CSV (missing header)")
print("=" * 80)
print(f"Shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())
```

```python
print("\n" + "=" * 80)
print("PROBLEM 2: Split Name -> Firstname + Lastname")
print("=" * 80)

df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True, n=1)

df = df.drop('Name', axis=1)

print("\nDataFrame sau split Name:")
print(df[['Id', 'Firstname', 'Lastname', 'Age', 'Weight']].head())
```

```python
print("\n" + "=" * 80)
print("PROBLEM 3: ChuáşŠn hĂła Weight - chuyáťn lbs -> kg")
print("=" * 80)

weight = df['Weight']

for i in range(len(weight)):
    x = str(weight[i])

    if "lbs" in x[-3:]:
        x = x[:-3]
        float_x = float(x)
        y = int(float_x / 2.2)
        y = str(y) + "kg"
        weight[i] = y

    elif "kg" in x[-2:]:
        weight[i] = x

    elif x == 'nan' or x == '-':
        weight[i] = np.nan

df['Weight'] = weight

print("\nWeight values sau chuáşŠn hĂła (chuyáťn lbs -> kg):")
print(df[['Id', 'Firstname', 'Lastname', 'Weight']].head(10))
```

```python
print("\n" + "=" * 80)
print("PROBLEM 4: XĂła rows ráťng (NaN)")
print("=" * 80)

print(f"\nBefore dropna: {len(df)} rows")

df.dropna(how="all", inplace=True)

print(f"After dropna: {len(df)} rows")
print("\nDataFrame sau xĂła rows ráťng:")
print(df)
```

```python
print("\n" + "=" * 80)
print("PROBLEM 5: Drop duplicates (dáťąa trĂŞn subset columns)")
print("=" * 80)

print(f"\nBefore drop duplicates: {len(df)} rows")

# Drop duplicates dáťąa trĂŞn subset: Firstname, Lastname, Age, Weight
df = df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])

print(f"After drop duplicates: {len(df)} rows")
print("\nDataFrame sau xĂła duplicates:")
print(df)
```

```python
print("\n" + "=" * 80)
print("PROBLEM 6: XĂła non-ASCII characters")
print("=" * 80)

df['Firstname'] = df['Firstname'].replace({r'[^\x00-\x7F]+': ''}, regex=True)
df['Lastname'] = df['Lastname'].replace({r'[^\x00-\x7F]+': ''}, regex=True)

print("\nFirstname & Lastname sau xĂła non-ASCII:")
print(df[['Firstname', 'Lastname']])
```

```python
print("\n" + "=" * 80)
print("PROBLEM 7: Missing Values - CĂĄc chiáşżn lĆ°áťŁc xáť­ lĂ˝")
print("=" * 80)

print("\nđ Missing Values Statistics:")
print(df[['Age', 'Weight']].isnull().sum())

print("\nCĂĄc chiáşżn lĆ°áťŁc xáť­ lĂ˝ missing values:")
print("  a. Deletion: XĂła records cĂł missing values")
print("  b. Dummy substitution: Thay báşąng dummy value (e.g., 0)")
print("  c. Mean substitution: Thay báşąng mean")
print("  d. Frequent substitution: Thay báşąng giĂĄ tráť pháť biáşżn nháşĽt")
print("  e. Improve the data collector: LiĂŞn háť clients")

print("\nâ Cháťn chiáşżn lĆ°áťŁc: Mean substitution + Deletion (náşżu thiáşżu cáşŁ 2)")

df['Weight_numeric'] = df['Weight'].str.extract(r'(\d+)').astype(float)

age_mean = df['Age'].mean()
weight_mean = df['Weight_numeric'].mean()

print(f"\nAge mean: {age_mean:.2f}")
print(f"Weight mean: {weight_mean:.2f}")

print("\nâď¸ Xáť­ lĂ˝ missing values:")
print(f" - Rows thiáşżu Age: {df['Age'].isnull().sum()}")
print(f" - Rows thiáşżu Weight: {df['Weight'].isnull().sum()}")

missing_both = df[(df['Age'].isnull()) & (df['Weight'].isnull())]
print(f" - Rows thiáşżu cáşŁ Age vĂ  Weight: {len(missing_both)}")

df = df.drop(missing_both.index)

df['Age'] = df['Age'].fillna(age_mean)

df['Weight_numeric'] = df['Weight_numeric'].fillna(weight_mean)
df['Weight'] = df['Weight_numeric'].astype(int).astype(str) + "kg"

df = df.drop('Weight_numeric', axis=1)

print("\nDataFrame sau xáť­ lĂ˝ missing values:")
print(df[['Id', 'Firstname', 'Lastname', 'Age', 'Weight']])
```

```python
print("\n" + "=" * 80)
print("PROBLEM 8: Unpivot dáťŻ liáťu - tĂĄch heart rate columns")
print("=" * 80)

print("\nBefore unpivot shape:", df.shape)

id_vars = ['Id', 'Age', 'Weight', 'Firstname', 'Lastname']
value_vars = ['m0006', 'm0612', 'm1218', 'f0006', 'f1218'] # Removed 'f0612' as it doesn't exist

df_melted = pd.melt(
    df[id_vars + value_vars],
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='sex_and_time',
    value_name='PulseRate'
)

print("\nAfter melt shape:", df_melted.shape)

def extract_sex_and_time(sex_and_time_str):
    """Extract Sex vĂ  Time táťŤ sex_and_time"""

    sex = sex_and_time_str[0]
    sex_full = 'male' if sex == 'm' else 'female'

    time_code = sex_and_time_str[1:]

    # Ensure time_code has enough characters before slicing
    if len(time_code) >= 4:
        hours_lower = int(time_code[:2])
        hours_upper = int(time_code[2:4])
    elif len(time_code) == 3: # Handle cases like '006' (m006, f006)
        hours_lower = int(time_code[0]) # Single digit for lower bound
        hours_upper = int(time_code[1:3]) # Two digits for upper bound
    else: # Fallback for unexpected formats
        hours_lower = 0
        hours_upper = 0

    time_range = f"{hours_lower:02d}-{hours_upper:02d}"

    return sex, sex_full, time_range

sex_data = df_melted['sex_and_time'].apply(extract_sex_and_time)

df_melted['Sex_code'] = sex_data.apply(lambda x: x[0])
df_melted['Sex'] = sex_data.apply(lambda x: x[1])
df_melted['Time'] = sex_data.apply(lambda x: x[2])

df_melted = df_melted.drop('sex_and_time', axis=1)

print("\nColumns sau extract:")
print(df_melted.columns.tolist())

print("\nFirst 10 rows sau unpivot:")
print(df_melted.head(10))
```
