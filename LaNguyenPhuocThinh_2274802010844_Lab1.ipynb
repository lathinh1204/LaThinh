```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col="STT")
df.head(10)
```

```python
df.tail(10)
```

```python
df['DT'].fillna(0, inplace = True)
df
```

```python
df['T1'].fillna(df['T1'].mean(), inplace = True)
df
```

```python
score_cols = [c for c in df.columns
              if c not in ['STT', 'GT', 'DT', 'KV', 'KT', 'T1', 'DH1', 'DH2', 'DH3']]

for col in score_cols:
    df[col] = df[col].fillna(df[col].mean())
df
```

```python
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] +
              df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10

df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] +
              df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10

df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] +
              df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10
print(df[['TBM1', 'TBM2', 'TBM3']].head())
df
```

```python
def xep_loai(tbm):
    if tbm < 5.0:
        return 'Y'
    elif tbm < 6.5:
        return 'TB'
    elif tbm < 8.0:
        return 'K'
    elif tbm < 9.0:
        return 'G'
    else:
        return 'XS'

df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)
print("XL1:", df['XL1'].value_counts().to_dict())
print("XL2:", df['XL2'].value_counts().to_dict())
print("XL3:", df['XL3'].value_counts().to_dict())
df
```

```python
for tbm, us in [('TBM1', 'US_TBM1'), ('TBM2', 'US_TBM2'), ('TBM3', 'US_TBM3')]:
    min_val = df[tbm].min()
    max_val = df[tbm].max()
    df[us] = ((df[tbm] - min_val) / (max_val - min_val)) * 4

print(df[['US_TBM1', 'US_TBM2', 'US_TBM3']].head().round(3))
```

```python
def ket_qua_xet_tuyen(row):
    kt = row['KT']
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']
    if kt == 'A1':
        score = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B':
        score = (dh1 + dh2*2 + dh3) / 4
    else:
        score = (dh1 + dh2 + dh3) / 3
    return 1 if score >= 5.0 else 0

df['KQXT'] = df.apply(ket_qua_xet_tuyen, axis=1)
print(df['KQXT'].value_counts().rename({1: 'Đậu', 0: 'Rớt'}).to_dict())
df
```
